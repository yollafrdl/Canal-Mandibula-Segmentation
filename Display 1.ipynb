```python
import numpy as np
import matplotlib.pyplot as plt
import os
import glob
import cv2
```

```python
path = 'Data/10/Sagital/200.bmp'
gt_path = 'Data/10/GT Sagital/200.jpg'
```

```python
img2 = cv2.imread(path, 0)
plt.imshow(img2, cmap='gray')
```

```python
gt = cv2.imread(gt_path, 0)
plt.imshow(gt, cmap='gray')
```

```python
edges = cv2.Canny(gt,100,200)
plt.imshow(edges, cmap='gray')
```

```python
def show(img, gt):
    ret, thresh = cv2.threshold(gt, 127, 255, 0)
    contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    img_gt = cv2.drawContours(img, contours, -1, (255, 255, 255), -1)
    return img_gt
```

```python
hasil = show(img2, gt)
```

```python
plt.figure(figsize=(16, 20))
plt.subplot(1,3,3)
plt.imshow(hasil, cmap='gray')
plt.title('Image + GT'), plt.xticks([]), plt.yticks([])
plt.show()
```
